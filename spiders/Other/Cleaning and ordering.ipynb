```python
import pandas as pd
import numpy as np
import xlsxwriter
from difflib import SequenceMatcher
similarity_ratio = 0

def similar(a, b):
    global similarity_ratio
    similarity_ratio = SequenceMatcher(None, a, b).ratio()
    return similarity_ratio
```
